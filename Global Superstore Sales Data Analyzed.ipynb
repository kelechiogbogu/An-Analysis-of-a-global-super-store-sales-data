**QUESTION 4:Which products do our customers order the most?**

```python
TopTenProducts = global_superstore.groupby('Product_Name')[['Quantity']].sum().sort_values(by = 'Quantity', ascending = False).head(10)
TopTenProducts
```

**QUESTION 5:Were there any losses recorded??**

```python
no_profit = global_superstore.loc[global_superstore['Profit'] < 0] 
no_profit.head(5)
```

**QUESTION 6:If yes, which products recorded the most losses?**

```python
products_losses = no_profit.groupby('Product_Name')[['Profit']].sum().sort_values(by = 'Profit', ascending = True).head(10)
products_losses
```

**QUESTION 7:Which Products generated the highest profits?**

```python
#To filter out years where losses and zero profit were made.
sales_profit = global_superstore.loc[global_superstore['Profit'] > 0] 
sales_profit.head(5)
```

```python
products_profit = sales_profit.groupby('Product_Name')[['Profit']].sum().sort_values(by = 'Profit', ascending = False).head(10)
products_profit
```

**QUESTION 8:Compare the percentage of transactions that resulted in profits, no profits and losses.**

```python
#To answer this question, I defined a variable that categorizes each transaction as either Profit, No Profit or Loss.
def transaction_result(Profit):
    """Classifies a transaction as either Profit, No Profit or Loss"""
    if Profit < 0:
        return "Loss"
    elif Profit == 0:
        return "No Profit"
    else:
        return "Profit"
    
#Applying the defined function
global_superstore['Transaction_Outcome']=global_superstore['Profit'].apply(transaction_result)
global_superstore.head(5)
```

### pd.DataFrame()
This is used to convert data into a tabular format.
After classifying each row based on the transaction outcome, the next step is to group them together using the count() function.

```python
transaction_outcome = global_superstore.Transaction_Outcome.value_counts()
pd.DataFrame(transaction_outcome)
```

**Showing Distribution of order across countries**

**In order to show the distribution of sales across countries, I need to visualize the result data in a map.**

**Note that when ploting a map, it is best to make use of the three-letter country code associated with each country, this ensures that a more reliable result is achieved because the three-letter country code is equivalent to the ISO-3 which is part of the geometry used to plot the map.**

**Because this data does not contain any info about the 3-letter code, I have to retrieve it from the COUNTRY columnusing pycountry**

```python
import pycountry
```

```python
# Defining a function to that will be extract the code from the Country column
def get_country_code (country_name):
    try:
        return pycountry.countries.get(name=country_name).alpha_3
    except:
        return ("NA")
    
# Applying the function    
Sales_by_Countries['Country_Code'] = Sales_by_Countries.apply(lambda row: get_country_code(row.Country), axis = 1)
Sales_by_Countries.head()
```
